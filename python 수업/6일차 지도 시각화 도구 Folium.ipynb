## 미국의 실업률 지도에 표시해 보기

```python
import pandas as pd
state_unpl = pd.read_csv("C:/Users/user/Desktop/imsi/data/06. US_Unemployment.csv")
state_unpl
```

```python
state_geoInfo = "C:/Users/user/Desktop/imsi/data/06. US_Unemployment.json"
map = fo.Map(location=[40,-98],zoom_start = 4)
map.choropleth(geo_data=state_geoInfo, data = state_unpl, columns=['State', 'Unemployment'], key_on='feature.id',fill_color='YlGn',
legend_name = 'Unemployment Rate(%)')
map
```

## 서울시 구별 데이터로 경계 표시하고, 살인 발생 비율 표시하기

```python
import json
seoul_geo_path = 'C:/Users/user/Desktop/imsi/data/07. Seoul_gu.json'
seoul_geo = json.load(open(seoul_geo_path, encoding='utf-8'))
crime_anal = pd.read_csv("C:/Users/user/Desktop/imsi/data/06. crime_anal.csv",encoding = "utf-8", index_col = '구별')
crime_anal.head()
```
