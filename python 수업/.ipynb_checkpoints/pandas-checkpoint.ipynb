```python
plt.figure
plt.scatter(dr['인구수'],dr['소계'],s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
fp1=np.polyfit(dr['인구수'],dr['소계'],1)
fp1
```

```python
f1=np.poly1d(fp1)
fx=np.linspace(100000,700000,100)
```

```python
plt.figure
plt.scatter(dr['인구수'],dr['소계'],s=5)
plt.plot(fx,f1(fx),ls='dashed',lw=3,color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
dr['오차']=np.abs(dr['소계']-f1(dr['인구수']))
df_sort=dr.sort_values(by='오차',ascending=False)
df_sort.head()
```
