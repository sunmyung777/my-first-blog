```python
import numpy as np
```

```python
x=np.arange(-7,7,0.01)
y=1*np.sin(1*(x-0))+0
plt.figure()
plt.plot(x,y,label='sin',color='red',lw=10)
plt.plot(x,np.cos(x),label='cos',color='blue',lw=10)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.axvline(x=0, color = 'black')
plt.axhline(y=0, color = 'r')
plt.grid()
plt.title('example')
plt.show()
```
