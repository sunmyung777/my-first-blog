## - webpage를 읽어오기 위해 urllib.request 모듈을 사용한다.

## - webpage의 데이터를 가져오기 위해 Beautiful Soup을 사용한다.

```python
import urllib.request as ur
```

## 웹페이지 읽어오기

```python
res = ur.urlopen('https://naver.com')
web = res.read().decode('utf-8')
web
```

## beautiful soup 가져오기

```python
import bs4
```

```python
bs = bs4.BeautifulSoup(web, 'html.parser')
head = bs.head
head
```

```python
body = bs.body
body
```

## 웹페이지 데이터에 접근하기

## 네이버에서 실시간 검색어 가져오기

```python
rank = bs.find_all('li', class_='ah_item')
rank
```

```python
rank[0]
```

```python
rank[0].find('span',class_ = "ah_k").getText()
```

### Q. 네이버 실시간 검색어 1~10위 가져와서 출력해 보자
